```python
import pandas as pd
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
```

```python
file_path = './Hotel Reservations.csv'
hotel_data = pd.read_csv(file_path)
```

```python
hotel_data.head()
```

```python
hotel_data.describe()
```

```python
target = 'booking_status'
features = hotel_data.columns.drop([target, 'Booking_ID'])
```

```python
features
```

```python
categorical_cols = hotel_data[features].select_dtypes(include=['object']).columns
numerical_cols = hotel_data[features].select_dtypes(include=['int64', 'float64']).columns

numerical_cols
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('numerical', StandardScaler(), numerical_cols),
        ('categorical', OneHotEncoder(), categorical_cols)
    ])
```

```python
preprocessor
```

```python
X = hotel_data[features]
y = hotel_data[target]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train_transformed = preprocessor.fit_transform(X_train)
X_test_transformed = preprocessor.transform(X_test)
```

```python
log_reg = LogisticRegression(max_iter=1000, random_state=42)
log_reg.fit(X_train_transformed, y_train)
```

```python
decision_tree = DecisionTreeClassifier(random_state=42)
decision_tree.fit(X_train_transformed, y_train)
```
